Now we are finished with the numeric. Lets deal with the non numeric

A good predictor will have a high variability so i will make a metrics for Sale Price median range(max median-min median) for each categories. This will help me know what non numeric to priortize first.

```python
def median_range(column):
    max_median = df.groupby(column)['SalePrice'].median().max()
    min_median = df.groupby(column)['SalePrice'].median().min()
    median_range = max_median - min_median
    return median_range
```

Let's Create a dictionary for the ranges of non_numeric_column

```python
non_numeric_column = df.select_dtypes(exclude = np.number).columns.tolist() #create columns for non numeric
```

Using a for loop lets compute the median range and store it in a dictionary feature range

```python
feature_ranges ={}
for column in non_numeric_column:
    rng = median_range(column)
    feature_ranges[column] = rng

non_numeric_column_corr = pd.Series(feature_ranges).sort_values(ascending = False)
non_numeric_column_corr
```

## 4. Data Preparation and Feature Engineering

**Goal:** To transform our raw, cleaned data into a fully-numeric and optimized set of features suitable for machine learning. This crucial phase involves creating new, powerful predictors and converting all existing features into a format our models can understand.

Our strategy is divided into three main parts:

**a. Feature Engineering**
We **will create** new features from the existing data to provide more predictive power to our models:
* **TotalSF**: A feature representing the total square footage of the house by combining the basement, first, and second-floor areas.
* **HouseAge**: A feature representing the age of the house at the time of sale, calculated by subtracting 'YearBuilt' from 'YrSold'.
We will also measure the correlation of our features so that we can now handle our **repeating outliers**

**b. Feature Preparation & Encoding**
To finalize the dataset for modeling, we **will perform** the following steps:
* **Ordinal Encoding:** For categorical features with a clear, intrinsic rank (e.g., 'ExterQual', 'GarageCond'), we will map their text labels to ordered numbers. This allows the model to understand the inherent quality ranking.
* **Nominal Encoding:** For categorical features with no clear order (e.g., 'Neighborhood'), we will use One-Hot Encoding (pd.get_dummies).This is necessary because machine learning algorithms cannot process string data directly and would misinterpret a simple numerical mapping as an ordinal relationship.
* **Final Cleanup:** We will drop the 'Id' column as it serves only as an identifier and has no predictive value.

**c. Target Variable Transformation**
Based on our EDA, we identified a significant right-skew in the 'SalePrice' distribution. A heavily skewed target can violate the assumptions of linear regression models. To mitigate this, we **will apply** a log transformation (np.log1p) to normalize the distribution, which helps in building a more robust and reliable model.

Now this is better than the original skewed distribution

**Ordinal Feature Encoding**

For our categorical features that have a clear, intrinsic rank or order (e.g., quality ratings like **ExterQual** where **Excellent** > **Good** > **Fair**), we cannot use one-hot encoding without losing this valuable ranking information. At the same time, machine learning models require numerical input.

To solve this, we will use **Ordinal Encoding**. This technique involves mapping each ranked category to a specific number that preserves its inherent order (e.g., **{'Fair': 1, 'TA': 2, 'Gd': 3, 'Ex': 4}**). By converting these labels into an ordered numerical format, we allow our model to understand and leverage the ranking of these features, which can significantly improve its predictive performance.

**Nominal Feature Encoding**

For our remaining categorical features that have no intrinsic rank or order (e.g., 'Neighborhood', 'MSZoning'), a simple numerical mapping would be inappropriate. The model would incorrectly assume a ranked relationship between the categories (e.g., that one neighborhood is "greater than" another).

To solve this, we will use **One-Hot Encoding**. This technique converts each category within a feature into a new binary column (containing 1s or 0s). This allows the model to understand the influence of each individual category without imposing a false sense of order.

We will implement this using the **pd.get_dummies()** function from the pandas library. To prevent multicollinearity (where features are perfectly predictable from each other), we will set the **drop_first=True** parameter. This removes the first category for each feature, as its presence can be inferred from the other categories, ensuring our data is statistically sound for modeling.

```python

df=pd.get_dummies(df, drop_first=True)
```

```python
df.info()
```

### c. Target Variable Transformation

A critical assumption for many regression models, particularly Linear Regression, is that the model's errors are normally distributed. In our initial Exploratory Data Analysis (EDA), we observed that the `SalePrice` target variable was heavily **right-skewed**.

To address this and help our models perform more reliably, we will apply a log transformation to the **SalePrice** column. This mathematical transformation will reduce the skewness of the distribution, making it more symmetric and closer to a normal distribution. For numerical stability, we will use the **np.log1p** function, which calculates **log(1 + x)**.

```python
df['SalePrice']=np.log1p(df['SalePrice'])
```

## 5. Model Training & Evaluation

**Goal:** To build, evaluate, and compare several regression models using our prepared data to identify the one that most accurately and reliably predicts `SalePrice`. Our approach is to start with a simple baseline and progressively test more complex models, using hyperparameter tuning to find the optimal performer.

Our modeling workflow will be as follows:

* **1. Data Splitting:** First, we will separate our prepared dataset into features (`X`) and our transformed target (`y`). We will then split this data into training and testing sets using `train_test_split`. This is a critical step to ensure we have an "unseen" test set for an unbiased evaluation of our models' true performance.
* **2. Model Training & Tuning:** We will train and tune three different models to compare their effectiveness:
    * **Linear Regression:** To serve as a simple, interpretable baseline.
    * **Random Forest:** An ensemble model to see if we can improve on the baseline.
    * **XGBoost:** A powerful gradient boosting model that we will tune to achieve the best possible performance.
* **3. Final Evaluation & Selection:** We will compare the final R-squared and RMSE scores of all models on both the training and testing sets. This will allow us to check for overfitting and confidently select our final "champion" model.

First, We need to make a dataframe that will store the performance of our 3 models (The model name will be in the rows/index and the columns will be the train/test performance metrics

```python
# Make a list for the indices
model_names = ['Linear Regression', 'Tuned Random Forest', 'Tuned XGBoost']

# Below are the data tahat will populate the column  for the each metrics and datasets

# For R2 Scores
train_r2_scores = [train_r2_lr, train_r2_rfr, train_r2_xgb] 
test_r2_scores = [r2_score_lr, r2_score_rfr, r2_score_xgb] 

# For  RMSE Scores
train_rmse_scores = [train_rmse_lr, train_rmse_rfr, train_rmse_xgb]
test_rmse_scores = [rmse_score_lr, rmse_score_rfr, rmse_score_xgb]

# For MSE Scores
train_mse_scores = [train_mse_lr, train_mse_rfr, train_mse_xgb]
test_mse_scores = [mse_score_lr, mse_score_rfr, mse_score_xgb]
```

Let's store the list of  data for each metrics in a dictionary called summary_data

```python
summary_data = {
    'Train r2 Score': train_r2_scores,
    'Test r2 Score': test_r2_scores,
    'Train RMSE': train_rmse_scores,
    'Test RMSE': test_rmse_scores,
    'Train MSE': train_mse_scores,
    'Test MSE': test_mse_scores
}


```

Now we can make our data frame using pandas and make the model name as our indeces

```python
summary_df = pd.DataFrame(summary_data, index=model_names)

summary_df
```

Now that we made our data frame lets make the the visualization. The strategy is to create a  1x4 subplot grid for our dashboard. The first one will contain our main take aways and the 2nd, 3rd and 4th will contain the R2 score, RMSE and MSE score respctively.

```python
fig, axs = plt.subplots(1, 4, figsize=(28, 7), constrained_layout=True)
fig.suptitle('Final Model Performance Dashboard', fontsize=24)


axs[0].axis('off') # Turn off the axis for the text block
summary_text = (
    f"Key Takeaways:\n\n"
    f"1. Best Model: The tuned XGBoost model is the champion,\n"
    f"   achieving the highest Test r2 Score ({summary_df.loc['Tuned XGBoost']['Test r2 Score']:.4f}).\n\n"
    f"2. Overfitting: The Random Forest model shows clear\n"
    f"   signs of overfitting.\n\n"
    f"3. Baseline: The simple Linear Regression provided a\n"
    f"   very strong and reliable baseline performance."
)
axs[0].text(0.01, 0.5, summary_text, ha='left', va='center', fontsize=14, wrap=True)


model_names = summary_df.index
x = np.arange(len(model_names))
width = 0.35


ax1 = axs[1]
ax1.bar(x - width/2, summary_df['Train r2 Score'], width, label='Train Score', color='lightgray')
ax1.bar(x + width/2, summary_df['Test r2 Score'], width, label='Test Score', color='dodgerblue')
ax1.set_title('Train vs. Test: r2 Score', fontsize=16)
ax1.set_ylim(0.85, 1.0)


ax2 = axs[2]
ax2.bar(x - width/2, summary_df['Train RMSE'], width, label='Train Score', color='lightgray')
ax2.bar(x + width/2, summary_df['Test RMSE'], width, label='Test Score', color='dodgerblue')
ax2.set_title('Train vs. Test: RMSE', fontsize=16)
ax2.set_ylim(0.0, 0.20)


ax3 = axs[3]
ax3.bar(x - width/2, summary_df['Train MSE'], width, label='Train Score', color='lightgray')
ax3.bar(x + width/2, summary_df['Test MSE'], width, label='Test Score', color='dodgerblue')
ax3.set_title('Train vs. Test: MSE', fontsize=16)
ax3.set_ylim(0.0, 0.03)


for ax in [ax1, ax2, ax3]:
    ax.set_xticks(x, model_names, fontsize=11)
    ax.legend(frameon=False)
    
   
    ax.bar_label(ax.containers[0], fmt='{:.4f}', padding=-18, color='gray', fontsize=10)
    ax.bar_label(ax.containers[1], fmt='{:.4f}', padding=-18, color='white', fontsize=10)

  
    ax.spines[['top', 'right', 'bottom', 'left']].set_visible(False)
    ax.tick_params(axis='y', left=False, labelleft=False)
    ax.tick_params(axis='x', bottom=False)

plt.show()
```
